```python
# STEP 6: Fetch products from API

url = "https://dummyjson.com/products?limit=20"

try:
    response = requests.get(url, timeout=5)

    if response.status_code == 200:
        data = response.json()
        products = data["products"]

        print("ID | Title | Category | Price | Rating")
        print("--------------------------------------------------")

        for item in products:
            print(f"{item['id']} | {item['title']} | {item['category']} | ${item['price']} | {item['rating']}")

    else:
        print("Error: Received status code", response.status_code)

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Unexpected error:", e)
```

```python
# STEP 7: Filter and sort products

# taking only products with rating >= 4.5
filtered_products = []

for item in products:
    if item["rating"] >= 4.5:
        filtered_products.append(item)

# sorting by price (highest to lowest)
filtered_products.sort(key=lambda x: x["price"], reverse=True)

print("\nFiltered and Sorted Products:\n")

for item in filtered_products:
    print(f"{item['title']} | Price: ${item['price']} | Rating: {item['rating']}")
```

```python
# STEP 10: Safe divide function

def safe_divide(a, b):
    try:
        result = a / b
        return result

    except ZeroDivisionError:
        return "Error: Cannot divide by zero"

    except TypeError:
        return "Error: Invalid input types"


# testing the function
print(safe_divide(10, 2))     # normal
print(safe_divide(10, 0))     # zero division
print(safe_divide("ten", 2))  # wrong type
```

```python
# STEP 11: Safe file reader

def read_file_safe(filename):
    try:
        file = open(filename, "r", encoding="utf-8")
        content = file.read()
        file.close()
        return content

    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")

    finally:
        print("File operation attempt complete.")


# testing
print(read_file_safe("python_notes.txt"))   # should work
print(read_file_safe("ghost_file.txt"))     # should fail
```

```python
# STEP 12: API call with proper exception handling

url = "https://dummyjson.com/products?limit=5"

try:
    response = requests.get(url, timeout=5)

    if response.status_code == 200:
        data = response.json()
        print("\nProducts fetched successfully:\n")

        for item in data["products"]:
            print(item["title"], "| Price:", item["price"])

    else:
        print("Error: Received status code", response.status_code)

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Unexpected error:", e)
```

```python
# STEP 13: Input validation loop

while True:
    user_input = input("\nEnter product ID (1–100) or 'quit': ")

    if user_input.lower() == "quit":
        print("Exiting...")
        break

    # check if number
    if not user_input.isdigit():
        print("Invalid input. Please enter a number.")
        continue

    product_id = int(user_input)

    # check range
    if product_id < 1 or product_id > 100:
        print("Please enter ID between 1 and 100.")
        continue

    # API call
    try:
        url = f"https://dummyjson.com/products/{product_id}"
        response = requests.get(url, timeout=5)

        if response.status_code == 404:
            print("Product not found.")

        elif response.status_code == 200:
            data = response.json()
            print("Title:", data["title"])
            print("Price:", data["price"])

        else:
            print("Error: Received status code", response.status_code)

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")

    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")

    except Exception as e:
        print("Unexpected error:", e)
```

```python
# STEP 14: Logging function

from datetime import datetime

def log_error(function_name, error_message):
    file = open("error_log.txt", "a", encoding="utf-8")

    # formatted timestamp
    time_now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    log = f"[{time_now}] ERROR in {function_name}: {error_message}\n"

    file.write(log)
    file.close()
```

```python
# STEP 15: Trigger errors and log them

# 1. Connection Error (fake URL)
try:
    bad_url = "https://this-host-does-not-exist-xyz.com/api"
    requests.get(bad_url, timeout=5)

except requests.exceptions.ConnectionError as e:
    print("Connection error triggered")
    log_error("fetch_products", f"ConnectionError - {e}")


# 2. HTTP 404 error (manual check)
url = "https://dummyjson.com/products/999"

try:
    response = requests.get(url, timeout=5)

    if response.status_code != 200:
        print("404 error triggered")
        log_error("lookup_product", f"HTTPError - {response.status_code} Not Found for product ID 999")

except requests.exceptions.ConnectionError as e:
    log_error("lookup_product", f"ConnectionError - {e}")

except requests.exceptions.Timeout as e:
    log_error("lookup_product", f"Timeout - {e}")

except Exception as e:
    log_error("lookup_product", str(e))


# STEP 16: Read and print log file

file = open("error_log.txt", "r", encoding="utf-8")

print("\nError Logs:\n")
print(file.read())

file.close()
```
